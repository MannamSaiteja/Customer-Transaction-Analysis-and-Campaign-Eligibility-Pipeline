```python
import pandas as pd
import numpy as np
```

```python
#load customers data
uci_customers=pd.read_csv('./Datasets/bank-full.csv', delimiter=';')
uci_customers.head()
```

```python
#load transactions data
kaggle_transactions=pd.read_csv('./Datasets/bank_transactions.csv')
kaggle_transactions.head()
```

```python
# data transformation for kaggle_transactions
kaggle_transactions = kaggle_transactions.rename(columns={
    "TransactionID": "transaction_id",
    "CustomerID": "customer_id",
    "TransactionDate": "transaction_date",
    "TransactionAmount (INR)": "amount"
})

kaggle_transactions["transaction_date"] = pd.to_datetime(kaggle_transactions["transaction_date"], format="%d/%m/%y").dt.date

kaggle_transactions["transaction_type"] = np.random.choice(["credit", "debit"], size=len(kaggle_transactions))

kaggle_transactions = kaggle_transactions[["customer_id", "transaction_date", "transaction_type", "amount"]]

# Save the formatted transactions file
kaggle_transactions.to_csv("formatted_transactions.csv", index=False)
```

```python
transactions = pd.read_csv("formatted_transactions.csv")
unique_customer_ids = transactions["customer_id"].unique()

# Ensure the customers dataset has enough rows by repeating UCI customers
num_customers_needed = len(unique_customer_ids)
uci_customers_expanded = pd.concat([uci_customers] * (num_customers_needed // len(uci_customers) + 1), ignore_index=True)

uci_customers_expanded = uci_customers_expanded.head(num_customers_needed)

uci_customers_expanded.rename(columns={
    "age": "age",
    "job": "job",
    "marital": "marital_status",
    "balance": "account_balance"
}, inplace=True)

# Assign customer IDs from transactions
uci_customers_expanded.insert(0, "customer_id", unique_customer_ids)

# Generate random credit scores between 450-850
uci_customers_expanded["credit_score"] = np.random.randint(450, 850, size=num_customers_needed)

uci_customers_expanded = uci_customers_expanded[["customer_id", "age", "job", "marital_status", "account_balance", "credit_score"]]

# Save the formatted dataset
uci_customers_expanded.to_csv("formatted_customers.csv", index=False)

```
